```python
%matplotlib inline
import pandas
import numpy as np
import dao
from matplotlib import pyplot as plt

df = pandas.read_csv('d.csv', index_col=0)
d,p,r,_ = dao.get_dpr(df)
latest = np.load('lasttime.npy')
```

```python
from datetime import datetime
timestamp = int(df.columns[-1])/ 1000
print(datetime.fromtimestamp(timestamp))
timestamp = int(df.columns[0])/ 1000
print(datetime.fromtimestamp(timestamp))
```

```python
import yfinance

def getStocks(select):
    return list(map(lambda x: x.split('#')[1], df.index[select]))

def testSelect(select, start, end):
    stocks = getStocks(select)
    s = yfinance.download(stocks, start=start, end=end)['Close'].T
    d, p2, r, _ = dao.get_dpr(s)
    p2 = p[select]
    s = (d.T * p2).sum(axis=1)
    s = s/s[0] * 10
    return s, p2
```

```python
ss = []
for i in range(100):
    s, _ = testSelect(latest[i][0], "2015-03-08", "2019-01-01")
    drop, meandrop, a, fitloss = dao.stat(s)
    ss.append([i,drop,meandrop,a,fitloss])
```

```python
ana = pandas.DataFrame(ss)
ana[ana[1] < 30].sort_values(4).iloc[:30,:]
#ana.sort_values(4).iloc[:30,:]
```

```python
ana = pandas.DataFrame(ss)
best = ana[ana[1] < 1.5].sort_values(4).iloc[:30,:]
best = latest[best[0].values.tolist()][:,0]
ss2 = []
for i in best:
    s,_ = testSelect(i, "2015-03-08", "2019-01-01")
    s = pandas.DataFrame(s)
    s.columns = [str(i)]
    ss2.append(s)
print(len(ss2))
s = pandas.concat(ss2, axis=1)
s.plot(figsize=(32,18))
```
